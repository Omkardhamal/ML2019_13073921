##Aim:
The motive of this report is to develop a system that can detect failure of Air-conditioning systems before they breakdown. If we have a look at the current scenario in the world of air-conditioning asset maintenance, most of the maintenance that is undertaken is reactive. Reactive maintenance means repairing an asset after it breaks down. Maintaining assets reactively leads to increase in capital expenditure and loss of time and inconvenience. The number of breakdowns increases with reactive maintenance. Hence, to overcome this issue a method to maintain assets pro-actively (before they breakdown, so that the breakdown does not occur) is proposed. It is expected that analysing the health of these assets remotely and detecting any issues beforehand will give rise to efficiency and productivity of an organisation.

##Background
####Assumption
The clients are industrial clients and not residential. 
As of today, 90% of the maintenance work carried out is reactive i.e. when an asset or an equipment breaks down the client calls up the entity responsible for looking after the asset. The entity arrives and repairs the equipment. The problem with Reactive maintenance is that the repair costs are five to ten times more than the cost of preventive or proactive maintenance. The clients have no idea about how the air-conditioning unit looks from inside, they ignore its health as they can’t see them because they are hidden inside the ducts or walls. This makes it difficult for them to understand what the problem is and they prefer reactive maintenance. But if they are shown how unhealthy the assets look from inside, they will be shocked and won’t realise how bad the condition was of the Air-Conditioning unit. For making the assets visible to the clients they need to be shown photographs of the equipment. It is not possible to convert all the reactive asset maintenance to pro-active but at least it can be reduced from 90% to 20%. Also, when an asset is about to breakdown, it operates abruptly, this means that it consumes more energy (KWh) than it actually requires. (Avonda, 2019)

If maintenances are scheduled depending on their current condition, then there will be less consumer complaints and the clients can concentrate on more important aspects of their life. On successful implementation of this idea it is believed that the customers standard of living will increase as they won’t have to worry about unnecessary things which can be eliminated by using data analytics in Pro-active air conditioning asset maintenance.
